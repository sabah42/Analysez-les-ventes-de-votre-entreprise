## Importer les packages

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

## Importer les données

```python
df_vente = pd.DataFrame(pd.read_csv("transactions.csv"))
df_produit=pd.DataFrame(pd.read_csv("products.csv"))
df_client= pd.DataFrame(pd.read_csv("customers.csv"))
```

## Analyse de la forme des données

```python
df_vente.dtypes
df_client.dtypes
df_produit.dtypes
```

```python
df_produit['categ']=df_produit['categ'].astype('object')# convertire le type pour ne pas confodus entre les variables qualitatives et les variables quantitatives 
```

### Analyse des valeurs aberrantes

```python
df_produit.describe()
#l'existance des valeurs aberrantes: le prix min est négatif
```

```python
df_produit=df_produit[df_produit['price']>0]# éliminer les lignes ou le prix est négatif
df_produit.shape
```

```python
df_vente.describe()
#l'existance des valeurs aberrantes: la colonne date contient les dates des tests
```

```python
df_vente[df_vente['date'].str.find('test')==0]
```

```python
df_vente=df_vente[df_vente['date'].str.find('test')!=0]# éliminer les lignes ou la colonne date contient les dates des tests
df_vente
```

```python
df_vente.loc[(df_vente['client_id']=='ct_0')&(df_vente['client_id']=='ct_1')]
```

```python
df_client.sort_values(by=['client_id'], ascending=[True])
# #l'existance des valeurs aberrantes: les deux dernières lignes
```

```python
df_client=df_client.loc[(df_client['client_id']!='ct_0')&(df_client['client_id']!='ct_1')]# supprimer les valeurs aberrantes
df_client.shape
```

```python
df_produit.loc[(df_produit['id_prod']=='T_0')]# vérifier l'existance de produit T_0 
```

### Analyse des valeurs manquantes

```python
df_ven_cl=pd.merge(df_vente,df_client,on='client_id',how= 'outer')
df=pd.merge(df_ven_cl,df_produit, on='id_prod',how='outer') # on a fait une jointure externe pour garder toute les informations
df
```

```python
df.isnull().sum().sort_values()# calculer le nbr des valeurs manquantes
```

```python
#les raisons de 124 valeurs manquantes de prix et de cat
dtn=df[(df['categ'].isna()==True)&(df['price'].isna()==True)] 
# le produit 0_2245 qui apparaît dans fichiers vente il est absent du catalogue des produits
dtn.head(103)
```

```python
dtn.tail(21) # 21 clients enregistrés n'ont acheté aucun produit
```

```python
#les raisons de 43 valeurs manquantes de date et de session_id
df[(df['date'].isna()==True)&(df['session_id'].isna()==True)] 
# 21 clients enregistrés n'ont acheté aucun produit et 22 produits n'ont pas encore vendu 
```

```python
df[(df['client_id'].isna()==True)&(df['sex'].isna()==True)&(df['birth'].isna()==True)]
#22 produits n'ont pas encore vendu
```

```python
df[(df['id_prod'].isna()==True)]
# 21 clients enregistrés n'ont acheté aucun produit
```

```python
df=df.dropna(axis=0).copy()#éliminer les lignes où il manque des données
# on a supprimé 146 lignes (103 ligne ou il y a le produit 0_2245 qui apparaît dans fichiers vente il est absent du catalogue des produits et 21 lignes ou il ya des clients enregistrés n'ont acheté aucun produit et 22 lignes ou il y'a des produits qui n'ont pas encore vendu)
```

```python
df.duplicated().sum()# vérifier les lignes répétées
```
